### `Hypothesis-01:` - Chi-Squared Test
### `H_0:` There is no significant difference in diesel and petrol cars sale in different cities.
### `H_1:` There is significant difference in diesel and petrol cars sale in different cities.

```python
# Create a contingency table
contingency_table = pd.crosstab(df['Fuel_Type'], df['Location'])

# print(contingency_table, "\n")

# Perform the Chi-Squared test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)
print("p_value = ", p_value , "\n")

# print the result by using if-else condition.
if p_value < 0.05:
    print("The sale of 'Petrol Cars' is significantly different to 'Diesel Cars' at all Locations. (reject H0)")
else:
    print("The sale of 'Petrol Cars' approximately equal to 'Diesel Cars' at all Locations. (fail to reject H0)")
```

### `Hypothesis-03:` - Z-test (two tail)
### `H_0:` The `Price` means of `Diesel and Petrol Vehicles` are significantly equal.
### `H_1:` The `Price` means of `Diesel and Petrol Vehicles` are significantly unequal.

+ **First of all, check the normality of the price columns of diesel and petrol cars.**

```python
diesel_prices = df[df['Fuel_Type'] == 'Diesel']['Price_lac']
petrol_prices = df[df['Fuel_Type'] == 'Petrol']['Price_lac']

# Perform the Shapiro-Wilk test to check the normality of the data.
statistic, p = stats.shapiro(diesel_prices)

# Print the results using if-else condition.
print('p-value =', p , '\n')
if p > 0.05:
    print('The "Diesel Cars" prices are Normally Distributed.\n')
else:
    print('The "Diesel Cars" prices are not Normally Distibuted.\n')

print("----------------------------------------")

# Perform the Shapiro-Wilk test to check the normality of the data.
statistic, p = stats.shapiro(petrol_prices)

# Print the results using if-else condition.
print('p-value =', p , '\n')
if p > 0.05:
    print('The "Petrol Cars" prices are Normally Distributed.\n')
else:
    print('The "Petrol Cars" prices are not Normally Distibuted.\n')
```

+ **Double check the normaility by using QQ-Plot.**

```python
# Create a figure with subplots
plt.figure(figsize=(12, 4))

# Probability plot for diesel prices
plt.subplot(1, 2, 1)
stats.probplot(diesel_prices, dist="norm", plot=plt)
plt.title('Probability Plot - Diesel Prices')

# Probability plot for petrol prices
plt.subplot(1, 2, 2)
stats.probplot(petrol_prices, dist="norm", plot=plt)
plt.title('Probability Plot - Petrol Prices')

plt.tight_layout()
plt.show()
```

+ **Applying two-tail `Z-test` to test the hypothesis as the data is `approximately normal`.**

```python
# Calculate sample means and standard deviations
mean_diesel = np.mean(diesel_prices)
std_diesel = np.std(diesel_prices, ddof=1)
n_diesel = len(diesel_prices)

mean_petrol = np.mean(petrol_prices)
std_petrol = np.std(petrol_prices, ddof=1)
n_petrol = len(petrol_prices)

# Perform z-test
z_value = (mean_diesel - mean_petrol) / np.sqrt((std_diesel**2 / n_diesel) + (std_petrol**2 / n_petrol))

# Set significance level (alpha)
alpha = 0.05

# Calculate critical z-value for a two-tailed test
critical_z_value = stats.norm.ppf(1 - alpha/2)

# Compare z-value to critical z-value
if np.abs(z_value) > critical_z_value:
    print("There is a significant difference in mean prices of diesel and petrol cars. (reject H_0)")
else:
    print("There is no significant difference in mean prices of diesel and petrol cars. (fail to reject H_0)")
```

+ **Test the `Homogeniety` of the data columns `diesel_prices` and `petrol_prices`.**

+ **`Levene's Test`**

```python
# Perfom the levene's test on the 'diesel_prices' and 'petrol_prices' data
stat, p = stats.levene(diesel_prices, petrol_prices)
print("Levene's test p-value:", p)

# print the result based on if-else condition
if p > 0.05:
    print("The variances of 'diesel_prices' and 'petrol_prices' are equal.")
else:
    print("The variances of 'diesel_prices' and 'petrol_prices' are not equal.")
```

+ **Apply `One-Way ANOVA` Test.**
  
+ **Here, the variance of both `diesel_prices` and `petrol_prices` are not equal.**

+ **Imp. Note:** Without fullfilling the Assumptions, the results of ANOVA test are not trustworthy.

```python
# Apply One-Way ANOVA Test
f_statistic, p_value = stats.f_oneway(diesel_prices, petrol_prices)

print("p_value = ", p_value , "\n")

# print the result based on if-else condition
if p_value < 0.05:
    print("The prices of diesel and petrol cars are significantly different. (reject H_0)")
else:
    print("The prices of diesel and petrol cars are not significantly different. (fail to reject H_0)")
```

+ **There is only one pair `diesel-petrol`, So, no need to apply `Post-hoc Test` for further analysis.**

### `Hypothesis-04:` - t-test
### `H_0:` The `Mileage` means of `Diesel and Petrol Vehicles` are significantly equal.
### `H_1:` The `Mileage` means of `Diesel and Petrol Vehicles` are significantly unequal.

+ **First of all, check the `normality` of the data.**

```python
diesel_mileage = df[df['Fuel_Type'] == 'Diesel']['Mileage']
petrol_mileage = df[df['Fuel_Type'] == 'Petrol']['Mileage']

# Perform the Kolmogorov-Smirnov Test to check the normality of the data.
statistic, p = stats.kstest(diesel_prices, 'norm')

# Print the results using if-else condition.
print('p-value =', p , '\n')
if p > 0.05:
    print('The "Diesel Cars" mileage values are Normally Distributed.\n')
else:
    print('The "Diesel Cars" mileage values are not Normally Distibuted.\n')

print("-------------------------------------------------------")

# Perform the Kolmogorov-Smirnov Test to check the normality of the data.
statistic, p = stats.kstest(petrol_prices, 'norm')

# Print the results using if-else condition.
print('p-value =', p , '\n')
if p > 0.05:
    print('The "Petrol Cars" mileage values are Normally Distributed.\n')
else:
    print('The "Petrol Cars" mileage values are not Normally Distibuted.\n')
```

+ Double check the `normality` test by ploting the `QQ-Plots`.

```python
# Create a figure with subplots
plt.figure(figsize=(12, 4))

# Probability plot for diesel prices
plt.subplot(1, 2, 1)
stats.probplot(diesel_mileage, dist="norm", plot=plt)
plt.title('Probability Plot - Diesel Prices')

# Probability plot for petrol prices
plt.subplot(1, 2, 2)
stats.probplot(petrol_mileage, dist="norm", plot=plt)
plt.title('Probability Plot - Petrol Prices')

plt.tight_layout()
plt.show()
```

+ **Finally, Plot the `Histograms` to verify the above results of `normality`.**

```python
# Set the style (optional)
sns.set(style="whitegrid")

# Create a figure and axes with subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6), sharey=True)

# Plot the histograms
sns.histplot(diesel_mileage, bins=30, kde=True, ax=axes[0], color='blue', label='Diesel Cars')
sns.histplot(petrol_mileage, bins=30, kde=True, ax=axes[1], color='orange', label='Petrol Cars')

# Set labels and title
axes[0].set_xlabel('Mileage')
axes[1].set_xlabel('Mileage')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Histogram of Mileage Values of "Diesel Cars"')
axes[1].set_title('Histogram of Mileage Values of "Petrol Cars"')

# Add legend
axes[0].legend()
axes[1].legend()

# Show the plot
plt.tight_layout()
plt.show()
```

+ Both distributions are already approximately normal.

**Applying two-tail `t-test for two independent samples.`**

```python
# Perform t-test for two independent samples.
t_statistic, p_value = stats.ttest_ind(diesel_mileage, petrol_mileage, equal_var=False)

print("p-value = ", p_value , "\n")

# Access the statistical significance.
if p_value < 0.05:
    print("There is a significant difference in mean values of mileage of diesel and petrol cars. (reject H_0)")
else:
    print("There is no significant difference in mean values of mileage of diesel and petrol cars. (fail to reject H_0)")
```

+ **Test the `Homogeniety` of the data columns `diesel_mileage` and `petrol_mileage`.**

+ **`Bartlett's Test`**

```python
# Perfom the levene's test on the 'diesel_prices' and 'petrol_prices' data
stat, p = stats.bartlett(diesel_mileage, petrol_mileage)
print("Bartlett's test p-value:", p , "\n")

# print the result based on if-else condition
if p > 0.05:
    print("The variances of 'diesel_mileage' and 'petrol_mileage' are equal.")
else:
    print("The variances of 'diesel_mileage' and 'petrol_mileage' are not equal.")
```

+ **Apply `One-Way ANOVA` Test.**
  
+ **Here, the variance of both `diesel_mileage` and `petrol_mileage` are not equal.**

+ **Imp. Note:** Without fullfilling the Assumptions, the results of ANOVA test are not trustworthy.

```python
# Apply One-Way ANOVA Test
f_statistic, p_value = stats.f_oneway(diesel_mileage, petrol_mileage)

print("p_value = ", p_value , "\n")

# print the result based on if-else condition
if p_value < 0.05:
    print("The mileage of diesel and petrol cars are significantly different. (reject H_0)")
else:
    print("The mileage of diesel and petrol cars are not significantly different. (fail to reject H_0)")
```

### `Hypothesis-05:` - Chi-Squared Test
### `H_0:` The annual sales of the `number of diesel cars` are equal to the annual sales of `number of petrol cars`.
### `H_1:` The annual sales of `number of diesel cars` are not equal to the annual sales of `number of petrol cars`.

```python
# Create a contingency table
contingency_table = pd.crosstab(df['Year'], df['Fuel_Type'])

# print(contingency_table, "\n")

# Perform the Chi-Squared test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)
print("p_value = ", p_value , "\n")

# print the result by using if-else condition.
if p_value > 0.05:
    print("The number of diesel and petrol cars soled annually are significantely equal. (fail to reject H0)")
else:
    print("The number of diesel and petrol cars soled annually are significantely different. (reject H0)")
```
